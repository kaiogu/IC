## Exercícios de otimização 1D

### Exe. 1

Use o otimizador de gradiente descendente (`tf.train.GradientDescentOptimizer`) para achar o mínimo de uma função de segunda ordem da forma `y = a*x^2 + b*x + c`, sendo `a` positivo. O loop deve imprimir o numero da iteração (step_number) o valor de `y` e de `x` para cada iteração.




```python
tf.reset_default_graph()

starting_val = tf.placeholder(tf.float64, shape=[])
a, b, c = (0.5, 2, -4)
x = tf.Variable(starting_val)
y = a*x**2 + b*x + c
optimizer = tf.train.GradientDescentOptimizer(0.)
train_op = optimizer.minimize(y)
init_op = tf.global_variables_initializer()
num_steps = 20

x_vals = np.empty([num_steps])
y_vals = np.empty([num_steps])

with tf.Session() as sess:
    sess.run(init_op, feed_dict={starting_val:5})
    for i in range(20):
        x_vals[i] = x.eval()
        y_vals[i] = y.eval()
        print('Step {:>2}: y = {:=5.3}  at  x = {:=5.3}'.format(i, y_vals[i], x_vals[i]))
        sess.run(train_op)

tf.summary.FileWriter('logs', tf.get_default_graph()).close()
```
